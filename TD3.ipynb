```python
import numpy as np
tab = np.arange(0, 1.01, 0.01)
def func(x):
    func = x**3.0 + 2.0*x -2.0
    return func

print("Valeur entre 0 et 1:" )
for x in tab:

    print(str(round(x,2)) + "  " + str(func(x)))
    




print("solution f(0) et f(0.5) :")
print("f(0)= "+ str(func(0)))
print("f(0.5)= "+ str(func(0.5)))
```
